```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
res = np.array([0])
```

```python
res = np.concatenate((res, np.random.rand(1000)))
res = np.concatenate((res, -1 * np.random.rand(500)))
res = np.concatenate((res, 1 * np.random.rand(300)))
res = np.concatenate((res, -1 * np.random.rand(400)))
res = np.concatenate((res, 1 * np.random.rand(700)))
res = np.concatenate((res, -1 * np.random.rand(500)))
res[-1] = 0
```

```python
plt.plot(res)
plt.ylabel('Hand Position')
plt.show()
```

```python
np.savetxt("tes_data.csv", res, delimiter=",")
```
